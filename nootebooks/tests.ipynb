```python
# dates 
from_date = '2017-01-04'
to_date = '2017-01-05'

# google keywords
kw_list = ['bitcoin']
# language
pytrends = TrendReq(hl='en-US')


# number of total days to iterate
days = (date.fromisoformat(to_date)-date.fromisoformat(from_date)).days+1

# getting data per dates
for day in range(0,days):
    iterating_on_date = date.fromisoformat(from_date)+timedelta(days=day)
    print('Getting data for '+iterating_on_date.isoformat())
    data = pytrends.get_historical_interest(kw_list, 
                                            year_start=iterating_on_date.year,
                                            month_start=iterating_on_date.month, 
                                            day_start=iterating_on_date.day,
                                            hour_start=0, 
                                            year_end=iterating_on_date.year, 
                                            month_end=iterating_on_date.month, 
                                            day_end=iterating_on_date.day, 
                                            hour_end=23, 
                                            cat=0, geo='', gprop='', sleep=0)
    data.reset_index(drop=False, inplace=True)
    print(data)

```
